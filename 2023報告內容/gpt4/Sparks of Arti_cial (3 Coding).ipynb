3.1.2 Real world scenarios

編碼挑戰可以評估算法和數據結構方面的技能。然而，它們往往不能全面反映現實世界編碼任務的複雜性和
多樣性，這需要專門的領域知識、創造力和對多個組件和庫的整合，以及改變現有代碼的能力。為了評估
GPT-4在更現實的環境中的編碼能力，我們設計了與數據可視化、LATEX編碼、前端開發和深度學習有關的
端到端真實世界編碼挑戰，每一項都需要不同的專業技能。對於每一項任務，我們為GPT-4提供高層次的指
令，要求它編寫用適當的語言和框架編寫代碼。在少數情況下，我們也會在代碼寫完後改變規範，要求更新。

**Interfacing with LATEX.**

對於計算機科學家和數學家來說，用LATEX寫作是一項重要的工作，但它的學習曲
線並不簡單。由於其嚴格的語法和缺乏良好的調試器，即使是專家也會犯一些惱人的錯誤，每天都要花幾個

**Executing pseudo-code.**

編譯和執行用編程語言編寫的代碼很容易，但這也要求嚴格遵守語法和語義。編譯器不能處 理模糊的或非正式的表達方式，或功能的自然語言描述。相比之下，我們要求GPT-4執行圖3.8中的非瑣碎 的偽代碼，注意到它能夠執行並解釋每一個步驟（包括遞歸）。 ChatGPT則不能執行，儘管它似乎能夠解 釋每一行代碼。在下面的例子中，GPT-4正確地解釋了合併數組函數的非正式描述，該函數將兩個數組合併為一個缺失 元素的數組。它還理解了以簡略方式定義的遞歸函數rec。值得注意的是，GPT-4直接執行了這些代碼，而 沒有將其翻譯成其他定義明確的編程語言。這證明了AGI模型作為自然語言編程的新型工具的潛力，它可以 徹底改變我們未來的編碼方式。
