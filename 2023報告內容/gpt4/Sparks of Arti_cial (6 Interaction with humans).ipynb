6.1 Understanding Humans: Theory of Mind

心智理論是將信仰、情緒、慾望、意圖和知識等心理狀態歸屬於自己和他人，並理解它們如何影響行為和交
流的能力[Wel92]。它包括反思別人的心理狀態的基本任務，以及反思別人對別人的心理狀態的反映的更高級
任務（等等）。回答 "愛麗絲相信什麼？"就需要前一種技能的例子，而回答 "鮑勃認為愛麗絲相信什麼？"就
需要後一種技能的例子。心智理論對於與其他智能體的有效溝通和合作至關重要，因為它允許人們推斷他們
的目標、偏好、動機和期望，並相應地調整自己的行動和話語。此外，心智理論對於向他人學習也很重要，
因為它使人能夠解釋他們的反饋、建議和示範。

6.2 Talking to Humans: Explainability

What makes an explanation good?

評價一個解釋的質量的一個可能的方法是檢查輸出一致性，也就是說，在輸入x和
上下文c的情況下，解釋是否與輸出y一致。在圖6.9中，我們將GPT-4與text-davinci-003進行對比，並註
意到後者產生的解釋與輸出不一致（因為它沒有解決字母Q的選擇問題）。
評價一個解釋的質量的另一個可能的方法是檢查它是否與GPT-4的模擬PT ，即它是否讓我們有能力對模
型在不同輸入（甚至不同背景）下的未來行為做出預測。我們稱之為過程一致性，這往往是人類對解釋的期
望或渴望，特別是當他們想要理解、調試或評估對系統的信任時。我們可以通過創建新的輸入來評估過程一
致性，在那裡解釋應該預測行為，如圖6.10所示（其中GPT-4是過程一致性的）。

What leads to process-consistency?

如果GPT-4對PT 的模擬很差，對不同輸入和背景下x或c的微小變化非常敏
感，那麼過程一致性就會被打破。在這種情況下，即使一個好的解釋過程PE ，用過程一致性來解釋PT ，也
不能充分解釋GPT-4對PT 的模擬。這種可變性也使得GPT-4對PE 的模擬更有可能發生變化，產生相互
衝突的解釋。有一種方法似乎有助於降低GPT-4對輸入的微小變化的敏感性，那就是詳細說明PT 是什麼（
通過有一個明確的背景，如圖6.7中的第二和第三節，或者最好更詳細）。
