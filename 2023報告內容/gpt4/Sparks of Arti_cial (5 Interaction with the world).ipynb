智能的一個關鍵方面是交互性，我們將其定義為與其他代理、工具和環境溝通並對其反饋做出反應的能力。
交互性對於智能是很重要的，因為它使代理人能夠獲得和應用知識，解決問題，適應不斷變化的情況，並實
現超出其個人能力的目標。例如，人類彼此之間以及與環境之間的互動，以合作、學習、教學、談判、創造
等。互動性要求代理人理解複雜的想法，快速學習，並從經驗中學習，因此它與我們對智能的定義緊密相連
。
在這一節中，我們探討了交互性的兩個維度：工具使用和具身交互。工具使用涉及到使用外部資源，如
搜索引擎、計算器或其他API，來完成對代理本身來說很難或不可能的任務。體現式交互涉及使用自然語言
作為文本界面，與模擬或現實世界的環境進行交互，並接受來自它們的反饋。

5.1.2 Discussion

本節中的例子表明，GPT-4既能自己識別和使用外部工具，以提高其性能。它能夠推理出它需要哪些工具

專門的培訓或微調。

我們現在註意到一些限制。首先，GPT-4仍然需要一個提示，說明它被允許或期望使用外部工具。在沒

5.2 Embodied Interaction

雖然工具的使用是交互性的一個重要方面，但現實世界中的大多數交互都不是通過API發生的。例如，人類
能夠使用自然語言與其他代理進行交流，探索和操縱他們的環境，並從他們行動的後果中學習。這種體現式
的互動要求代理理解背景、目標、行動和每輪互動的結果，並相應地進行調整。雖然GPT-4顯然不是具身的
，但我們探索它是否可以通過使用自然語言作為各種模擬或真實世界環境的文本界面來進行具身的互動。

5.2.1 熱身運動：地圖導航

在圖5.8中，我們準備了一張房子的 "地圖"，並要求GPT-4通過互動查詢來探索它。然後我們要求它用語言
和可視化的方式來描述它，並將其與真實的地圖進行比較。雖然GPT-4沒有探索整個房子，但它準確地
描述了它所探索的東西，儘管它的所有互動都是通過這個有限的文本界面進行的。

5.2.2 基於文本的遊戲

基於文本的遊戲對語言模型來說是一個自然的、具有挑戰性的領域，因為它們需要理解自然語言，推理遊戲
狀態，並生成有效的命令。基於文本的遊戲是一種交互式小說，其中一個代理通過自然語言描述和命令與環
境互動。代理人必須通過探索環境和操縱物體來完成一個給定的任務，如尋找寶藏或逃離地牢。我們首先測
試GPT-4是否能在基於文本的遊戲中探索環境以執行給定任務。在這個和下一個實驗中，我們使用
TextWorld [CKY +18]，一個用於生成和玩基於文本的遊戲的框架，來創建兩個具有不同給定任務的遊戲。
