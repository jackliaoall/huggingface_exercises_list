**漸進式任務**

這些任務可以以漸進或連續的方式解決，通過每次增加一個詞或句子，構成解決方向上的進展
。這些任務可以通過內容生成來解決，不需要任何重大的概念轉變或洞察力，而是依靠將現有的知識和技能
應用於給定的主題或問題。漸進式任務的例子是寫一篇文章的摘要，回答事實問題，根據給定的韻律方案創
作一首詩，或解決一個遵循標準程序的數學問題。

**不連續的任務**

這些任務的內容生成不能以漸進或連續的方式完成，而是需要某種 "尤里卡 "的想法，在解決
任務的過程中實現不連續的飛躍。內容的產生涉及到發現或發明一種看待或構築問題的新方法，從而使其餘
內容的產生成為可能。不連續任務的例子有：解決一個需要新穎或創造性地應用公式的數學問題，寫一個笑
話或謎語，提出一個科學假設或哲學論證，或創造一個新的流派或寫作風格。
