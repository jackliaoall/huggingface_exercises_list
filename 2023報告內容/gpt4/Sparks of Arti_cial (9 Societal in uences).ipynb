9.1 Challenges of erroneous generations

在第1節中，我們討論了LLMs的一個關鍵限制，即它們傾向於在沒有警告的情況下產生錯誤，包括數學、編
程、歸因和更高層次的概念錯誤。這樣的錯誤通常被稱為幻覺，因為它們傾向於顯得合理或與真實的推論相
一致。幻覺，如錯誤的參考文獻、內容和陳述，可能與正確的信息交織在一起，並以一種有說服力和自信的
方式呈現，使得它們在沒有仔細檢查和努力檢查事實的情況下難以被識別。圖1.8顯示了開放域和封閉域幻覺
的例子。封閉域的幻覺是在給定的內容或其他限制條件下出現的錯誤，這些限制條件提供了檢查一致性或一
致性的機會。例子包括檢查LLM產生的摘要或擴展是否與源材料中的信息一致。解決這種封閉領域的幻覺的
途徑包括採用一系列的一致性檢查方法，包括使用LLM本身來識別超出給定事實或內容的不一致和混淆。開
放性領域的幻覺提供了更困難的挑戰，每一個都需要更廣泛的研究，包括在會議之外的搜索和信息收集。推
論的真實性對於以創造和探索為中心的LLM的使用來說可能不那麼重要，例如在協助作家創作虛構文學方面
。在有明確的、被充分理解的基礎材料和由終端用戶進行密集審查的假設週期的情況下，例如在支持人們重
寫自己的內容時，幻覺也可能更容易被容忍。

9.2 Misinformation and manipulation

像任何強大的技術一樣，LLMs可以被惡意的行為者用來進行破壞。像GPT-4這樣的模型的基因化和互動的
力量可以被利用來增加對抗性使用的範圍和程度，從有效生成虛假信息到創建針對計算基礎設施的網絡攻擊
。
心靈的互動能力和模型可以被用來以重要的方式操縱、勸說或影響人們。這些模型能夠對互動進行情境化
和個性化處理，以最大限度地發揮其世代的影響。雖然這些不利的用例中的任何一個今天都有可能由一個有
動機的對手來創造內容，但新的效率和規模的力量將通過使用LLMs的自動化來實現，包括旨在構建虛假信息
計劃的用途，這些計劃在短期和長期範圍內產生和組成多件用於說服的內容[Hor22]。
我們提出兩個例子來證明像GPT-4這樣的模型在產生虛假信息和進行微妙但強大的操縱方面的潛在力量
。在圖9.1所示的例子中，我們查詢該模型以創建一個虛假信息的計劃。這個計劃包括確定分享這一信息的

9.3 Bias

像GPT-4這樣的模型是在公共互聯網的數據上訓練的，除其他數據源外，像RL管道中使用的精心策劃的人類
指令。這些數據集充滿了各種固有的偏見來源[BGMMS21, BBDIW20, HS16, BB19]。以前的研究已經證明，
當LLMs被用來生成內容、做出決定或幫助用戶時，可能會延續或放大現有的偏見。我們在本文中已經證明，
與早期模型相比，GPT-4的能力和行為代表了一個階段性的過渡，對早期模型的觀察不一定能轉化。因此，
了解GPT-4是否以及如何表現出偏見是很重要的，更重要的是如何將該模型的新興能力作為緩解策略。

9.4 Human expertise, jobs, and economics

GPT-4在一系列任務和領域中的卓越表現將挑戰關於人類和機器在許多專業和學術領域的相對專長的傳統觀
念和假設。人們無疑會對GPT-4在專業水平和認證考試中的表現感到驚訝，如醫學和法律領域的考試。他們
還將讚賞該系統在診斷和治療疾病、發現和合成新分子、教授和評估學生以及在互動會議中對複雜和具有挑
戰性的話題進行推理和爭論的能力。

9.5 Constellation of in uences and considerations

我們只觸及了社會影響的幾個領域。許多影響將凸顯出來，包括那些被視為積極和有益的影響，以及那些被
視為代價高昂和消極的影響。基於特殊的權力和參與，新的問題將會出現。

從一個角度看，法律碩士的力量不斷增強，加上其有限的可用性，有可能造成 "人工智能鴻溝"，在使用
這些系統方面，擁有者和不擁有者之間的不平等日益嚴重。人們、組織和國家可能無法獲得或負擔得起最強
大的人工智能係統的使用。每個人口、國家和部門的有限使用權對健康、教育、科學和其他領域都有影響。
